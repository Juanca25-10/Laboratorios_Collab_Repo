```python
student_grades = []

while True:
    print("\n--- Menú de Notas de Estudiantes ---")
    print("1. Ingresar nota de estudiante")
    print("2. Mostrar todas las notas y promedio") # Actualizado el texto
    print("3. Modificar nota de estudiante")
    print("4. Salir")

    choice = input("Seleccione una opción: ")

    if choice == '1':
        name = input("Ingrese el nombre del estudiante: ")
        while True:
            grade_input = input(f"Ingrese la nota de {name}: ")
            try:
                grade = float(grade_input)
                if 0 <= grade <= 100: # Assuming grades are between 0 and 100
                    student_grades.append({'name': name, 'grade': grade})
                    print(f"Nota de {name} ({grade}) ingresada exitosamente.")
                    break
                else:
                    print("Error: La nota debe estar entre 0 y 100.")
            except ValueError:
                print("Error: Por favor, ingrese un número válido para la nota.")
    elif choice == '2':
        if not student_grades:
            print("No hay notas ingresadas aún.")
        else:
            print("\n--- Notas de Estudiantes ---")
            total_grades = 0
            for student in student_grades:
                print(f"Nombre: {student['name']}, Nota: {student['grade']}")
                total_grades += student['grade']

            num_students = len(student_grades)
            try:
                average_grade = total_grades / num_students
                print(f"\nPromedio de notas: {average_grade:.2f}") # Muestra el promedio con 2 decimales
            except ZeroDivisionError:
                print("No se puede calcular el promedio porque no hay estudiantes ingresados.")
    elif choice == '3': # Lógica para modificar nota de estudiante
        attempts = 0
        max_attempts = 3
        student_found_and_modified = False

        while attempts < max_attempts and not student_found_and_modified:
            student_to_modify_name = input("Ingrese el nombre del estudiante cuya nota desea modificar: ")
            target_student = None
            for student in student_grades:
                if student['name'].lower() == student_to_modify_name.lower():
                    target_student = student
                    break

            if target_student:
                print(f"Estudiante '{target_student['name']}' encontrado. Nota actual: {target_student['grade']}")
                while True: # Loop para la entrada de la nueva nota
                    new_grade_input = input(f"Ingrese la nueva nota para {target_student['name']}: ")
                    try:
                        new_grade = float(new_grade_input)
                        if 0 <= new_grade <= 100:
                            target_student['grade'] = new_grade
                            print(f"Nota de {target_student['name']} actualizada a {new_grade}.")
                            student_found_and_modified = True
                            break # Sale del bucle de entrada de nota
                        else:
                            print("Error: La nota debe estar entre 0 y 100.")
                    except ValueError:
                        print("Error: Por favor, ingrese un número válido para la nota.")
            else:
                attempts += 1
                if attempts < max_attempts:
                    print(f"Estudiante '{student_to_modify_name}' no encontrado. Le quedan {max_attempts - attempts} intentos para encontrar al estudiante.")
                else:
                    print(f"Estudiante '{student_to_modify_name}' no encontrado. Ha agotado sus {max_attempts} intentos para modificar una nota.")
    elif choice == '4':
        print("Saliendo del programa. ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intente de nuevo.")
```

### Explicación del Archivo `script.js`

Este archivo JavaScript contiene la lógica interactiva de la página web, replicando las funciones de tu programa Python:

*   **`studentGrades` Array:** Un array global que almacena objetos de estudiantes, cada uno con un `name` y `grade`. Permite múltiples entradas con el mismo nombre, al igual que tu código Python original.
*   **`modifyAttempts` Objeto:** Rastrea los intentos de modificación por estudiante para replicar la lógica de 3 intentos del programa Python.
*   **`displayMessage(elementId, message, type)`:** Una función auxiliar para mostrar mensajes temporales de éxito, error o información al usuario en la interfaz.
*   **`clearGradesDisplay()`:** Restablece la visualización de la tabla de notas y el promedio.
*   **`add-grade-form` Event Listener:**
    *   Captura el envío del formulario para agregar notas.
    *   Realiza validaciones de entrada (nombre no vacío, nota entre 0 y 100).
    *   Agrega la nueva nota al array `studentGrades`.
    *   Actualiza la visualización de notas si ya está activa.
*   **`show-grades-button` Event Listener (`displayAllGradesAndAverage()`):**
    *   Se activa al hacer clic en el botón 'Mostrar Notas y Promedio'.
    *   Recorre el array `studentGrades` y rellena la tabla HTML con los datos.
    *   Calcula y muestra el promedio de todas las notas.
    *   Maneja el caso de que no haya notas ingresadas.
*   **`modify-grade-form` Event Listener:**
    *   Captura el envío del formulario para modificar notas.
    *   Realiza validaciones de entrada (nombre no vacío, nueva nota entre 0 y 100).
    *   Busca al estudiante por nombre (sin distinción entre mayúsculas y minúsculas).
    *   Si encuentra al estudiante, actualiza su nota.
    *   Implementa la lógica de `max_attempts` (3 intentos) para encontrar al estudiante, igual que en tu código Python.
    *   Muestra mensajes de éxito o error.
    *   Actualiza la visualización de notas si ya está activa.
*   **`reset-data-button` Event Listener:**
    *   Reinicia el array `studentGrades` a vacío.
    *   Borra los intentos de modificación.
    *   Limpia la tabla de notas y los campos de los formularios, simulando la opción 'Salir' y reseteando el estado del programa.

```python
%%writefile script.js
let studentGrades = [];
let modifyAttempts = {}; // Para rastrear los intentos de modificación por estudiante

// --- Funciones de ayuda ---
function displayMessage(elementId, message, type = 'info') {
    const messageElement = document.getElementById(elementId);
    messageElement.textContent = message;
    messageElement.className = `message ${type}`;
    messageElement.style.display = 'block';
    setTimeout(() => {
        messageElement.style.display = 'none';
        messageElement.textContent = '';
    }, 5000); // Ocultar el mensaje después de 5 segundos
}

function clearGradesDisplay() {
    document.getElementById('grades-table').style.display = 'none';
    document.getElementById('no-grades-message').style.display = 'block';
    document.querySelector('#grades-table tbody').innerHTML = '';
    document.getElementById('average-grade').textContent = '';
}

// --- Event Listeners ---
document.getElementById('add-grade-form').addEventListener('submit', function(event) {
    event.preventDefault(); // Prevenir el envío del formulario por defecto
    const nameInput = document.getElementById('add-student-name');
    const gradeInput = document.getElementById('add-student-grade');
    const name = nameInput.value.trim();
    const grade = parseFloat(gradeInput.value);

    if (!name) {
        displayMessage('add-message', 'El nombre del estudiante no puede estar vacío.', 'error');
        return;
    }
    if (isNaN(grade) || grade < 0 || grade > 100) {
        displayMessage('add-message', 'Error: La nota debe ser un número entre 0 y 100.', 'error');
        return;
    }

    // Comportamiento del Python original: permite múltiples entradas con el mismo nombre
    studentGrades.push({ name: name, grade: grade });
    displayMessage('add-message', `Nota de ${name} (${grade}) ingresada exitosamente.`, 'success');

    nameInput.value = '';
    gradeInput.value = '';
    // Opcionalmente, refrescar la visualización si 'mostrar notas' estaba activo
    if (document.getElementById('grades-table').style.display !== 'none') {
        displayAllGradesAndAverage();
    }
});

document.getElementById('show-grades-button').addEventListener('click', displayAllGradesAndAverage);

function displayAllGradesAndAverage() {
    const gradesTableBody = document.querySelector('#grades-table tbody');
    gradesTableBody.innerHTML = ''; // Limpiar entradas anteriores
    const noGradesMessage = document.getElementById('no-grades-message');
    const gradesTable = document.getElementById('grades-table');
    const averageGradeElement = document.getElementById('average-grade');

    if (studentGrades.length === 0) {
        noGradesMessage.style.display = 'block';
        gradesTable.style.display = 'none';
        averageGradeElement.textContent = 'No hay notas ingresadas aún.';
        averageGradeElement.className = 'message info';
        return;
    }

    noGradesMessage.style.display = 'none';
    gradesTable.style.display = 'table';
    let totalGrades = 0;

    studentGrades.forEach(student => {
        const row = gradesTableBody.insertRow();
        const nameCell = row.insertCell();
        const gradeCell = row.insertCell();
        nameCell.textContent = student.name;
        gradeCell.textContent = student.grade.toFixed(2); // Mostrar la nota con 2 decimales
        totalGrades += student.grade;
    });

    const average = totalGrades / studentGrades.length;
    averageGradeElement.textContent = `Promedio de notas: ${average.toFixed(2)}`;
    averageGradeElement.className = 'message success';
}

document.getElementById('modify-grade-form').addEventListener('submit', function(event) {
    event.preventDefault();
    const nameToModifyInput = document.getElementById('modify-student-name');
    const newGradeInput = document.getElementById('new-student-grade');
    const nameToModify = nameToModifyInput.value.trim();
    const newGrade = parseFloat(newGradeInput.value);

    if (!nameToModify) {
        displayMessage('modify-message', 'El nombre del estudiante a modificar no puede estar vacío.', 'error');
        return;
    }
    if (isNaN(newGrade) || newGrade < 0 || newGrade > 100) {
        displayMessage('modify-message', 'Error: La nueva nota debe ser un número entre 0 y 100.', 'error');
        return;
    }

    const maxAttempts = 3;
    // Inicializar o incrementar los intentos para el nombre específico (insensible a mayúsculas/minúsculas)
    const key = nameToModify.toLowerCase();
    modifyAttempts[key] = (modifyAttempts[key] || 0) + 1;
    let studentFound = false;

    if (modifyAttempts[key] <= maxAttempts) {
        for (let i = 0; i < studentGrades.length; i++) {
            if (studentGrades[i].name.toLowerCase() === key) {
                studentGrades[i].grade = newGrade;
                studentFound = true;
                displayMessage('modify-message', `Nota de ${studentGrades[i].name} actualizada a ${newGrade} exitosamente.`, 'success');
                delete modifyAttempts[key]; // Resetear intentos si se encuentra y modifica con éxito
                break;
            }
        }

        if (!studentFound) {
            const remainingAttempts = maxAttempts - modifyAttempts[key];
            if (remainingAttempts > 0) {
                displayMessage('modify-message', `Estudiante '${nameToModify}' no encontrado. Le quedan ${remainingAttempts} intentos para encontrar al estudiante.`, 'error');
            } else {
                displayMessage('modify-message', `Estudiante '${nameToModify}' no encontrado. Ha agotado sus ${maxAttempts} intentos para modificar una nota.`, 'error');
                delete modifyAttempts[key]; // Resetear intentos después de agotarlos
            }
        }
    } else {
        displayMessage('modify-message', `Estudiante '${nameToModify}' no encontrado. Ya ha agotado sus ${maxAttempts} intentos para modificar una nota.`, 'error');
        delete modifyAttempts[key]; // Asegurarse de que los intentos se reinicien después de exceder el límite
    }

    nameToModifyInput.value = '';
    newGradeInput.value = '';
    // Opcionalmente, refrescar la visualización si 'mostrar notas' estaba activo
    if (document.getElementById('grades-table').style.display !== 'none') {
        displayAllGradesAndAverage();
    }
});

document.getElementById('reset-data-button').addEventListener('click', function() {
    studentGrades = [];
    modifyAttempts = {}; // Limpiar también los intentos de modificación
    clearGradesDisplay();
    displayMessage('exit-message', 'Datos reiniciados. Saliendo del programa (simulado).', 'info');
    // Limpiar todos los campos de formulario
    document.getElementById('add-student-name').value = '';
    document.getElementById('add-student-grade').value = '';
    document.getElementById('modify-student-name').value = '';
    document.getElementById('new-student-grade').value = '';
});

// Configuración inicial de la visualización
clearGradesDisplay();
```
