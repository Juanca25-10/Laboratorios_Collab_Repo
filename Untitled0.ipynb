Actividad 1: Juego de adivinar el número
Pasos:
1. Importar el módulo random para generar números aleatorios.
2. Crear una función generar_numero() que retorne un número aleatorio entre 1 y 100.
3. Implementar un bucle while para el juego principal.
4. Dentro del bucle, usar try-except para manejar entradas no numéricas.
5. Usar if-elif-else para dar pistas al jugador (mayor, menor o correcto).
6. Implementar un contador de intentos.
7. Usar break para salir del bucle cuando se adivine el número.

```python
import random

def generar_numero():
    """Retorna un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

def juego_adivina_numero():
    numero_secreto = generar_numero()
    intentos = 0
    print("¡Bienvenido al juego de adivinar el número!")
    print("Estoy pensando en un número entre 1 y 100.")

    while True:
        try:
            entrada_usuario = input("Intenta adivinar el número: ")
            intento_jugador = int(entrada_usuario)
            intentos += 1

            if intento_jugador < numero_secreto:
                print("¡Demasiado bajo! Intenta con un número mayor.")
            elif intento_jugador > numero_secreto:
                print("¡Demasiado alto! Intenta con un número menor.")
            else:
                print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
                break
        except ValueError:
            print("Entrada inválida. Por favor, ingresa un número entero.")

# Para ejecutar el juego:
juego_adivina_numero()
```

## Actividad 2: Calculadora de operaciones básicas

```python
# 1. Definir funciones para cada operación
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ValueError("No se puede dividir por cero.")
    return a / b

def calculadora():
    print("¡Bienvenido a la calculadora de operaciones básicas!")

    # 2. Crear un menú de opciones usando un bucle while
    while True:
        print("\n--- Menú de Opciones ---")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Salir")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == '5':
            print("¡Hasta luego!")
            break # 6. Implementar una opción para salir del programa usando break

        if opcion not in ('1', '2', '3', '4'):
            print("Opción inválida. Por favor, selecciona un número entre 1 y 5.")
            continue

        # 4. Implementar try-except para manejar errores
        try:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))

            resultado = None
            # 3. Utilizar if-elif-else para seleccionar la operación
            if opcion == '1':
                resultado = suma(num1, num2)
                print(f"Resultado: {num1} + {num2} = {resultado}")
            elif opcion == '2':
                resultado = resta(num1, num2)
                print(f"Resultado: {num1} - {num2} = {resultado}")
            elif opcion == '3':
                resultado = multiplicacion(num1, num2)
                print(f"Resultado: {num1} * {num2} = {resultado}")
            elif opcion == '4':
                resultado = division(num1, num2)
                print(f"Resultado: {num1} / {num2} = {resultado}")

        except ValueError as e:
            print(f"Error: {e}")
        except Exception as e:
            print(f"Ocurrió un error inesperado: {e}")

        # 5. Usar continue para volver al menú principal después de cada operación
        # (El bucle while ya maneja esto, no se necesita un 'continue' explícito aquí a menos que quieras saltar el resto de la iteración en caso de error, pero el flujo actual es bueno)

# Para ejecutar la calculadora:
calculadora()
```

## Actividad 3: Lista de tareas

```python
# 1. Crear una lista vacía para almacenar las tareas.
tareas = []

# 2. Definir funciones para agregar, eliminar, y mostrar tareas.
def agregar_tarea():
    tarea = input("Ingresa la nueva tarea: ")
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea():
    mostrar_tareas()
    if not tareas:
        print("No hay tareas para eliminar.")
        return
    try:
        indice = int(input("Ingresa el número de la tarea a eliminar: ")) - 1
        if 0 <= indice < len(tareas):
            tarea_eliminada = tareas.pop(indice)
            print(f"Tarea '{tarea_eliminada}' eliminada.")
        else:
            print("Número de tarea inválido.")
    except ValueError:
        print("Entrada inválida. Por favor, ingresa un número.")

def mostrar_tareas():
    if not tareas:
        print("No hay tareas en la lista.")
        return
    print("\n--- Tus Tareas ---")
    # 5. Utilizar un bucle for para mostrar todas las tareas.
    for i, tarea in enumerate(tareas):
        print(f"{i + 1}. {tarea}")
    print("-------------------")

def lista_de_tareas():
    print("¡Bienvenido a tu Lista de Tareas!")
    # 3. Implementar un menú principal usando un bucle while.
    while True:
        print("\n--- Menú Principal ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")

        opcion = input("Selecciona una opción (1-4): ")

        # 4. Usar if-elif-else para manejar las opciones del menú.
        if opcion == '1':
            agregar_tarea()
        elif opcion == '2':
            # 6. Implementar try-except para manejar errores al eliminar tareas.
            eliminar_tarea()
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("¡Hasta luego!")
            # 7. Usar break para salir del programa.
            break
        else:
            print("Opción inválida. Por favor, selecciona un número entre 1 y 4.")

# Para ejecutar la lista de tareas:
lista_de_tareas()
```
