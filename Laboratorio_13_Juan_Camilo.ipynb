Actividad 1: Utiliza la librería TextBlob para analizar el sentimiento de varias frases. Debes
interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su
polaridad

```python
from textblob import TextBlob

frases = [
    "The weather today is absolutely beautiful!",
    "I am really disappointed with the new update.",
    "The event was well-organized but lacked excitement."
]

for frase in frases:
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity

    if polaridad > 0:
        sentimiento = "Positivo"
    elif polaridad < 0:
        sentimiento = "Negativo"
    else:
        sentimiento = "Neutral"

    print(f"Frase: {frase}")
    print(f"Polaridad: {polaridad}")
    print(f"Sentimiento: {sentimiento}")
    print("-" * 50)
```

Actividad 2: Entrena un modelo de clasificación de sentimientos utilizando Naive Bayes y
datos de ejemplo, y luego evalúa su precisión.

Frases para actividad 2:
i. "This app is very intuitive!"
ii. "I didn't enjoy the customer service."
iii. "Quite average experience."
iv. "Fantastic presentation!"
v. "The food was awful."


```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Datos (frases + etiquetas)
frases = [
    "This app is very intuitive!",
    "I didn't enjoy the customer service.",
    "Quite average experience.",
    "Fantastic presentation!",
    "The food was awful."
]

# Etiquetas: positivo, negativo, neutral
etiquetas = ["positivo", "negativo", "neutral", "positivo", "negativo"]

# Convertir texto a números
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(frases)

# Dividir datos (entrenamiento y prueba)
X_train, X_test, y_train, y_test = train_test_split(X, etiquetas, test_size=0.2, random_state=42)

# Modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X_train, y_train)

# Predicciones
y_pred = modelo.predict(X_test)

# Evaluación
precision = accuracy_score(y_test, y_pred)

print("Predicciones:", y_pred)
print("Valores reales:", y_test)
print("Precisión:", precision)
```
